```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks_df = pd.read_csv("C:/datasets/spotify/tracks.csv")
tracks_df.head()
```

```python

most_popular = tracks_df.query('popularity>90', inplace= False).sort_values('popularity', ascending= False)
most_popular[:10]
```

```python
tracks_df.set_index("release_date", inplace=True)
tracks_df.index = pd.to_datetime(tracks_df.index)
tracks_df.head()
```

```python
tracks_df[['artists']].iloc[0]
```

```python
tracks_df['duration'] = tracks_df['duration_ms'].apply(lambda x: round(x/1000))
tracks_df.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks_df['duration'].head()
```

```python
import pandas as pd

# Create a DataFrame with two columns
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Calculate the Pearson correlation coefficient between x and y
corr = df['x'].corr(df['y'], method='pearson')

print(corr)
```

```python
corr_df = tracks_df.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmax=1, center=0, cmap='inferno', linewidth=1, linecolor='black', cbar_kws={'orientation': 'vertical'},)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
heatmap.set_title('Correlation Heatmap Between Variables')
```

```python
tracks_df['dates'] = tracks_df.index.get_level_values('release_date')
tracks_df.dates=pd.to_datetime(tracks_df.dates)
years= tracks_df.dates.dt.year
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
ax = sns.histplot(years).set_title('Number of songs per year')
# Set title


# Show the plot
plt.show()
```

```python
total_dr = tracks_df.duration
fig_dims  = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x= years, y= total_dr, ax= ax, errwidth = False).set(title= "Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=  tracks_df.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="year vs Duration")
plt.xticks(rotation=60)
```
